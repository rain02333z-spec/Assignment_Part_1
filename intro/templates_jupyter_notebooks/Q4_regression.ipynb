# Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import seaborn as sns
```

```python
train_dataset_oh = pd.read_csv('../train_insurance_one_hot.csv')
x_train = train_dataset_oh.drop(columns=['id','charges','chargeGroup'], inplace=False)
y_train = train_dataset_oh[["charges"]]

test_dataset_oh = pd.read_csv('../test_insurance_one_hot.csv')
x_test = test_dataset_oh.drop(columns=['id','charges','chargeGroup'], inplace=False)
y_test = test_dataset_oh[["charges"]]
```

## a)

```python
sns.lmplot(
    data=train_dataset_oh,
    x="age",
    y="charges",
    height=6,
)

plt.title("Charges over Age with Linear Regression Line")
plt.show()
```

## b)

```python
y_train = train_dataset_oh["charges"]
classifier = LinearRegression()
# train the classifier
classifier.fit(x_train, y_train)
coef_df=pd.DataFrame({
    "feature": x_train.columns,
    "coefficient": classifier.coef_
})
display(coef_df)

predicted_train = classifier.predict(x_train)
MAE_train=mean_absolute_error(y_train, predicted_train)
print("Train MAE: ", MAE_train)
predicted_test = classifier.predict(x_test)
MAE_test=mean_absolute_error(y_test, predicted_test)
print("Test MAE: ", MAE_test)
```
